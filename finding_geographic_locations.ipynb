```python
import numpy as np
import pandas as pd

data = pd.read_json("./data/headline_cities_and_countries.json")
data = data.replace({None: np.nan})

data.head(10)
```

```python
from geonamescache import GeonamesCache
gc = GeonamesCache()

def latitude(row):
    cities = gc.get_cities_by_name(row['cities'])
    if not cities:
        return np.nan
    return list(cities[0].values())[0]['latitude']

def longitude(row):
    cities = gc.get_cities_by_name(row['cities'])
    if not cities:
        return np.nan
    return list(cities[0].values())[0]['longitude']

def countrycode(row):
    cities = gc.get_cities_by_name(row['cities'])
    if not cities:
        return np.nan
    return list(cities[0].values())[0]['countrycode']

data['latitude'] = data.apply(lambda row: latitude(row), axis=1) 
data['longitude'] = data.apply(lambda row: longitude(row), axis=1)
data['countrycode'] = data.apply(lambda row: countrycode(row), axis=1)

data.head(10)
```
